## example of short data 

```python
numberOfData  =  {"f1" : [12, 13 , 44  , 16 ,  76 , 123 ,  122 ,  80000]}
ndd  = pd.DataFrame(numberOfData)

def find_outliers(your_data) :
    Q1 = your_data.quantile(0.25)
    Q2 = your_data.quantile(0.50)
    Q3  =  your_data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1  -  1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return  (your_data < lower_bound) | (your_data > upper_bound)

find_outliers(ndd)

find_outliers(nd)
```

## example of short data : (none and replace) 

```python
d1 =pd.DataFrame({
    "A" : [None , 110 ,  112 ,  None, 125] , 
    "B" : ["police" ,  None , "ambuless" , "emergency" , "engineer"], 
    "C" : [None, None , None , "python" , "python"]
})
d1
```

```python
d1.isnull().sum() 
d1["C"].mode()[0]   ## output ==>  python
d1["C"].replace([np.nan] ,d1["C"].mode()[0] , inplace= True )
d1
```

```python
d1["A"].fillna(113 , inplace= True) ## type 2  :
d1
```

```python
newd1_1 = d1["C"].fillna( method = "ffill" )   ## try to read the top item 
newd1_2 = d1["C"].fillna(method = "bfill")     ## try to read the buttom item 
print(newd1_1)
print(newd1_2)
```

```python
##  type  ==> object ==> mode 
##  type ==> int , float ==>  mean 

d1 =pd.DataFrame({
    "A" : [None , 110 ,  112 ,  None, 125] , 
    "B" : ["police" ,  None , "ambuless" , "emergency" , "engineer"], 
    "C" : [None, None , None , "python" , "python"]
})
d1
```

```python
#d1.info()
```

```python
def fill_data (data_) :
    if data_.dtype == "object" :
        return data_.fillna(data_.mode()[0])
    else :
        return data_.fillna(data_.mean())
```

```python
d1.apply(fill_data)
```
