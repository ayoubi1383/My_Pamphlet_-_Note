```python
import matplotlib.pyplot as plt
import numpy as np
```

## plot 

```python
x = np.array([-3 , -2 , -1 , 0 , 1, 2 , 3])
y = np.array([ 9, 4, 1 , 0 , 1, 4, 9])
plt.figure(figsize= (6,4))
plt.plot(x,y , label = "xy" , color = "purple" , linewidth = 4 , linestyle = ":" , marker = "o" , markersize = 10
        , markeredgecolor = "blue" )
plt.title("my testing")
plt.xlabel("X" ,  c= "red" , size= 15)
plt.ylabel("Y" ,  c = "red" ,  size = 15)
plt.grid(c = "pink" ,  linestyle = "--" , linewidth = 0.5)
plt.legend()
# plt.savefig("plot.png")
plt.show()
```

```python
x = np.linspace(-20 , 20 , 20)
plt.figure(figsize=(6 ,4))
plt.plot(x ,x , label = "x" , linestyle = "--")
plt.plot(x ,  x*5 , label = "x*5")
plt.plot(x ,  x**2 , label = "x^2")
plt.legend()
plt.show()
```

## scatter plot 

```python
x= np.array([4 , 12 , 13 , 44 , 1 , 1000 , 333 , 245])
y= np.array([199 , 22 , 13 , 289 , -13 , 500 , 111 , 222])
size = np.arange(100 , 900 , 100)         ## amount of size for data (worth of the data)
c_range = np.random.choice(["red" , "purple" , "green"  ] , 8)   ###  set color for each one !!
plt.figure(figsize= (6,4))
plt.title("scatter plot !!" , size = 22 ,  c = "purple")
plt.scatter(x, y  , c= c_range , s= size )
plt.show()
```

## pie plot 

```python
x = np.array([10 ,30 , 60])
y = np.array(["A" , "B" , "C"])
explode_ = np.array([0.2 , 0 , 0 ])
plt.figure(figsize= (6, 4))
plt.pie( x  , labels = y , autopct =" %0.2f%%" , explode= explode_) 
plt.legend()
plt.show()
```

## bar plot 

```python
x = np.array(["A" , "B" , "C"])
y = np.array([10 ,30 , 60])
plt.figure(figsize= (6, 4))
plt.bar(x, y , width= 0.3)
plt.show()
```

```python
x = np.array(["Amir hossein Ayoubi" , "Taylor Swift" , "Hande Ercel"])
y = np.array([10 ,60 , 30])
plt.figure(figsize= (6, 4))
plt.barh(x, y , height= 0.3)
plt.show()
```

## hist plot 

```python
age  =  np.array([20 , 15, 22 , 19 , 25 , 60  ,  12,  10,  21  ,20 ,  43 , 80])
plt.figure(figsize=(6,4))
plt.hist(age)
plt.show()
```

```python
data  = np.random.normal(1, 100 , 1200)
plt.figure(figsize=(6,4))
plt.hist(data)
plt.show()
```

## boxplot   plot 

```python
data  = np.array([-2221 , -239 , 139,  -140 ,  -130 ,  125  , 300 , -98 ,  3 , 134 , 600 , -200])
plt.figure(figsize=(6,4))
plt.boxplot(data)
plt.show()
```

## subplot  plot 

```python
x= np.array([4 , 12 , 13 , 44 , 1 , 1000 , 333 , 245])
y= np.array([199 , 22 , 13 , 289 , -13 , 500 , 111 , 222])
plt.figure(figsize=(10, 6))

plt.subplot(2, 3 , 1)
plt.title("plot")
plt.plot(x, y )

plt.subplot(2, 3 , 2)
plt.title("scatter")
plt.scatter(x, y )

plt.subplot(2, 3 , 3)
plt.title("pie")
plt.pie(x )

plt.subplot(2, 3 , 4)
plt.title("bar")
plt.bar(x, y , width= 30 )

plt.subplot(2, 3 , 5)
plt.title("hist")
plt.hist( y )

plt.subplot(2, 3 ,6)
plt.title("boxplot")
plt.boxplot( x )

plt.show()
```
